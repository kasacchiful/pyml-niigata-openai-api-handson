```python
## 環境変数設定
import dotenv
dotenv.load_dotenv('./drive/MyDrive/openai.env')
```

```python
import os
import openai

## Configuration
openai.api_key = os.getenv("OPENAI_API_KEY")  # set OpenAI API Key
model_name = 'gpt-3.5-turbo'

## Message
question = 'Pythonの開発環境の構築方法を教えてください。'
messages = [
    {"role": "system", "content": "Pythonのエキスパートとして、お答えください。"},
    {"role": "user", "content": question},
]

## Run
# response = openai.ChatCompletion.create(
#     model = model_name,
#     messages = messages,
# )
## openai v1.0.0 以降 (2023/11/7 Release)
client = openai.OpenAI(
    ## defaults to os.environ.get("OPENAI_API_KEY") なので、
    ## dotenvで "OPENAI_API_KEY" で保存していれば、特に指定は不要
    api_key = os.getenv("OPENAI_API_KEY")
)
response = client.chat.completions.create(
    model = model_name,
    messages = messages,
)
```
